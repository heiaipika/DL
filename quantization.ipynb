```python
import torch
def linear_q_with_scale_and_zeropoint(
    tensor,scale,zeropoint,dtype=torch.int8):

    scaled_tensor=tensor/scale+zeropoint
    rounded_tensor=torch.round(scaled_tensor)
    q_min=torch.iinfo(dtype).min
    q_max=torch.iinfo(dtype).max
    q_tensor=rounded_tensor.clamp(q_min,q_max).to(dtype)

    return q_tensor
```

```python
test_tensor=torch.tensor(
    [[191.5,-19.9,730.4],
     [92.12,293.3,-187],
     [0,682.3,243.3]
    ]
)
scale=3.5
zeropoint=-70

quantized_tensor=linear_q_with_scale_and_zeropoint(test_tensor,scale,zeropoint)
quantized_tensor

dequantize_tensor=scale*(quantized_tensor.float()-zeropoint)
dequantize_tensor
```
